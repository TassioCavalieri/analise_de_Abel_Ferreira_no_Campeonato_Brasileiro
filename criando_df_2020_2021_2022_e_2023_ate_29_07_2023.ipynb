# Importando bibliotecas e dados iniciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados_entre_09_08_2020_e_29_07_2023.csv')
```

# Análises preliminares e realização do filtro para o jogos de Abel Ferreira

```python
dados.head()
```

```python
# obtendo df apenas do jogos do Palmeiras
palmeiras = dados.query('time_mandante == "Palmeiras" | time_visitante == "Palmeiras" ')
```

```python
palmeiras.sort_values(by = ['data'])
```

```python
df_palmeiras = palmeiras[['ano_campeonato', 'data', 'rodada', 'estadio', 'time_mandante', 'time_visitante', 'tecnico_mandante', 'tecnico_visitante', 'gols_mandante', 'gols_visitante', 'defesas_mandante', 'defesas_visitante', 'chutes_mandante', 'chutes_visitante']]
```

```python
df_palmeiras.info()
```

```python
df_palmeiras['data'] = pd.to_datetime(df_palmeiras['data'])
```

```python
df_palmeiras.info()
```

```python
df_palmeiras.head()
```

```python
df_palmeiras.sort_values(by = 'data', inplace= True)
```

```python
df_palmeiras_2020 = df_palmeiras.query('ano_campeonato == 2020')
df_palmeiras_2021 = df_palmeiras.query('ano_campeonato == 2021')
df_palmeiras_2022 = df_palmeiras.query('ano_campeonato == 2022')
df_palmeiras_2023 = df_palmeiras.query('ano_campeonato == 2023')
```
